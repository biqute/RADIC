Notebook to control the board

```python
# Imports
import time
from marcj import *
```

```python
!gcc FunctionPlayer.c -lasound -lm -o FunctionPlayer
!gcc FunctionReader.c -lasound -lm -o FunctionReader
```

```python
duration = 10
frequency = 500
type = 's'

nloops = 15
```

```python
# Setup of the board
player = FunctionPlayer()
player.set_duration(duration)
player.set_frequency(frequency)
player.set_waveform(type)
player.verbose()

reader = FunctionReader()
reader.set_nloops(nloops)
```

```python
# Play and read the signal
player.play()
time.sleep(duration/10)
reader.read()
```
